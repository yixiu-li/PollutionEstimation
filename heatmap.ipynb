### Plot heatmap for historical prediction values

```python
historical_data_path = r"C:\Users\yixiu\OneDrive\文档\GitHub\SuperCloud2\data_from_eosloan_reloc\historical_data\all_tiles_historical_resampled_500m\all_tiles_historical_resampled_500m_aod_resize_8bit_arcgis.csv"
historical_summary = pd.read_csv(historical_data_path)
historical_summary.head()
```

```python
prediction_his_path = r"C:\Users\yixiu\Relocation_Fieldwork\02_Data\3. Administrative\preds\1_input\standardized_yx_id120_v2_epoch6_500m_8bit_resize_255.csv"
prediction_his_data = pd.read_csv(prediction_his_path)
prediction_his_data = prediction_his_data.rename(columns={"img_name": "tile_path_8bit_arcgis"})
prediction_his_data.head()
```

```python
merged_his_df = pd.merge(historical_summary, prediction_his_data, on="tile_path_8bit_arcgis", how="left")
merged_his_df.head()
```

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt

# Convert the geometry column (WKT format) to spatial objects
merged_his_df["geometry"] = merged_his_df["geometry"].apply(wkt.loads)

# Convert to a GeoDataFrame
gdf = gpd.GeoDataFrame(merged_his_df, geometry="geometry")

# Check the data
print(gdf.head())
```

```python
gdf.columns
```

```python
unique_years = gdf["year"].unique()
print(unique_years)
```

```python
filtered_gdf_2006 = gdf[gdf["year"] == 2006]

import matplotlib.pyplot as plt

# Plot the filtered data
min_value, max_value = 60, 125
filtered_gdf_2006.plot(column="pred", cmap="YlOrRd", legend=True, figsize=(10, 8), vmin=min_value, vmax=max_value)

# Set the x and y axis limits
xmin, xmax = 76.90, 77.3
ymin, ymax = 28.45, 28.85
plt.xlim([xmin, xmax])
plt.ylim([ymin, ymax])

# Add a title
plt.title("Heatmap for Year 2006")
plt.show()
```
